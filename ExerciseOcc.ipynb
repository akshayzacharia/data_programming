### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",delimiter="|")
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean().round(2)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_ratio = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
sums = male_ratio[['F', 'M']].sum(axis=1)
male_ratio['MaleRatio'] = round( male_ratio['M'] / sums , 1)
male_ratio.sort_values(by='MaleRatio', ascending= False, inplace=True )
male_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_max = users.pivot_table(index='occupation', values=['age'], aggfunc={'age': [min, max]}, fill_value=0)
min_max
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age = users.pivot_table(index='occupation',columns=['gender'], values=['age'], aggfunc='mean', fill_value=0)
mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
percentages = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
sums = percentages[['F', 'M']].sum(axis=1)
percentages['MaleRatio%'] = round( 100*male_ratio['M'] / sums , 1)
percentages['FemaleRatio%'] = 100 - percentages['MaleRatio%']
percentages
```
